```python
dataset_path = "/content/drive/MyDrive/Data Set/vulnerabilities.csv"
```

```python
#Load Dataset
import pandas as pd
df = pd.read_csv(dataset_path)
display(df)
```

```python
#Check Dataset Information
df.info()
print("***************************************")
df.describe()
```

```python
#Check Missing Values
df.isnull().sum()

df.dropna(subset=["cwe"], inplace=True)
```

```python
#Remove Duplicate Records
df.duplicated().sum()
df.drop_duplicates(inplace=True)
#print(df)
```

```python
#Remove Unnecessary Spaces
text_columns = ['func','cwe','project','message']

for col in text_columns:
    df[col] = df[col].astype(str).str.strip()

display(df.head())
```

```python
#Convert Data Types
df['target'] = df['target'].astype(int)
display(df)
```

```python
#Check Unique Target Values
df['target'].value_counts()
```

```python
#Remove Special Characters from Text Columns
import re

def clean_text(text):
    text = re.sub(r'[^a-zA-Z0-9\s]', ' ', str(text))
    text = re.sub(r'\s+', ' ', text)
    return text.lower()

for col in ['func','cwe','project','message']:
    df[col] = df[col].apply(clean_text)
```

```python
#Encode Categorical Features
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['project'] = le.fit_transform(df['project'])
df['cwe'] = le.fit_transform(df['cwe'])
df['func'] = le.fit_transform(df['func'])

display(df)
```

```python
#Final Check
df.info()
df.isnull().sum()
```

```python
df_50k = df.sample(n=50000, random_state=42)

print(df_50k.shape)

df_50k.to_csv('/content/drive/MyDrive/Data Set/vulnerabilities_50k.csv', index=False)
display(df_50k)
```
